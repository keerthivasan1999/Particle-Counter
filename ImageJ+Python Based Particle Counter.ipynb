### Latest code for particle counting (6 cluster sizes)

After correcting the error in the counting of cluster having (>5)
Important: Include the file "Single_particle_area_measurement.py"

```python
import pandas as pd
import csv  
import statistics
import Single_particle_area_measurement #Importing the 'Single_particle_area_measurement.py'
import numpy as np
import matplotlib.pyplot as plt


#-------------------------------------------
#Finding the average area of particles
particle_average_area=Single_particle_area_measurement.Average_area()
print('Average area of the particle is',particle_average_area,'micrometer^2\n')
#-------------------------------------------

Image_Count=20
time=[0,60,120,180]
time_xn_total=[]
export_time_xn_total=[]
export_area_percent=[]
area_percent=[]
average_area_percent=0

#-------------------------------------------
#-------------------------------------------
#-------------------------------------------

for t in time:
    xn_total=[0,0,0,0,0,0]
    print(">> "+str(t)+"mins <<")
    for m in range(1,Image_Count+1,1):
        n=[0,0,0,0,0,0] #Cluster size
        xn=[0,0,0,0,0,0] #Number fraction of particles
        N=0 #Total no. of particles
        df = pd.read_csv(str(t)+"mins"+str(m)+".csv", header=0)
        
        #-------------------------------------------
        for i in range(len(df)):
            
            
            #-----------
            #This block is only for extracting the area percent from the 1st of of the "%Area" column
            #And this is only for t=0 mins because we are interested in finding the area percentage at t=0 mins
            #As t>0mins, area fraction decreases due to self assemblying of the particles
            if(t==0):
                if(df['%Area'][i]<10):
                    area_percent.append(df['%Area'][i])
                    average_area_percent+=df['%Area'][i]
            #-----------
            
            
            if(df['Area'][i]<1000):
            #Note: above 'if' statement is to avoid using the first row in each excel sheet.
            #First row is the "Area percentage" row, where the 'Area' will be more the 40000
                
                if df['Area'][i]>5 and df['Area'][i]<8:
                    n[0]=n[0]+1
                elif df['Area'][i]>10 and df['Area'][i]<14:
                    n[1]=n[1]+1
                elif df['Area'][i]>16 and df['Area'][i]<20:
                    n[2]=n[2]+1
                elif df['Area'][i]>21 and df['Area'][i]<25:
                    n[3]=n[3]+1
                elif df['Area'][i]>26 and df['Area'][i]<30:
                    n[4]=n[4]+1
                elif df['Area'][i]>30:
                    #print(df['Area'][i])
                    n[5]=n[5]+int(df['Area'][i]/particle_average_area)
        
        for q in range(0,6,1): #Loop for 6 cluster sizes
            if(q!=5):
                N=N+((q+1)*(n[q]))
            else:
                N=N+n[q]
        
        for k in range(0,5,1): #Loop for only 5 cluster sizes
            xn[k]=round(((k+1)*n[k]/N),2)
            xn_total[k]=xn_total[k]+xn[k]
        #print('Image number:',m,n,xn,N)
        
        for k in range(0,5,1): #Loop for the final cluster size (i.e, >5)
            xn[5]=xn[5]+xn[k]
            
        xn[5]=1-xn[5]
        xn_total[5]=xn_total[5]+xn[5]
        #-------------------------------------------
        
    
    #-------------------------------------------
    #Writing the area percentage of each image into a csv file
    #Ensure proper indentation, following block should be outside the 'm' loop
    if(t==0):
        m_dummy=1
        for dum in area_percent:
            time_area_percent=[]    
            header = ['Time (mins)', 'Image number', 'Area fraction']
            time_area_percent.append(t)
            time_area_percent.append(m_dummy)
            m_dummy+=1
            time_area_percent.append(dum)
            export_area_percent.append(time_area_percent)
            with open('area_percents.csv', 'w', encoding='UTF8', newline='') as l:
                writer = csv.writer(l)
                writer.writerow(header)
                writer.writerows(export_area_percent)
                writer.writerow(['Average',"=",average_area_percent/Image_Count])
                writer.writerow(['Standard deviation',"=",statistics.stdev(area_percent)])
     #-------------------------------------------

    

    #-------------------------------------------
    #Finding the average number fractions
    for k in range(0,6,1): #Loop for 6 cluster sizes
        xn_total[k]=round(xn_total[k]/Image_Count,2)
    print('\nAverage number fraction of each cluster are',xn_total,'\n\n')
     
            
    #Writing the number fraction details of each time into a csv file    
    time_xn_total=[]    
    header1 = ['Time (mins)', 'xn (n=1)', 'xn (n=2)', 'xn (n=3)', 'xn (n=4)', 'xn (n=5)', 'xn (n>5)']
    time_xn_total.append(t)
    time_xn_total.extend(xn_total)
    export_time_xn_total.append(time_xn_total)
    with open('Final_results.csv', 'w', encoding='UTF8', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(header1)
        writer.writerows(export_time_xn_total)
    #-------------------------------------------
        
print('Average area percentage =',round(average_area_percent/Image_Count,2))
print('Standard deviation =',round(statistics.stdev(area_percent),2))    
#-------------------------------------------
#-------------------------------------------
#-------------------------------------------




#@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
#Plotting the results available in the "Final_results.csv" csv file
df = pd.read_csv ("Final_results.csv", usecols = ['xn (n=1)','xn (n=2)','xn (n=3)','xn (n=4)','xn (n=5)','xn (n>5)'])

# set width of bar
barWidth = 0.15
fig = plt.subplots(figsize =(12, 8))
 
# set height of bar
#zero = [12, 30, 1, 8, 22, 10]
#sixty = [28, 6, 16, 5, 10, 12]
#onetwenty = [29, 3, 24, 25, 17, 18]
#oneeighty= [10, 21, 4, 14, 11, 12]
zero=[]
sixty=[]
onetwenty=[]
oneeighty=[]
for j in range(0,4,1):
    for i in range(0,6,1):
        if(j==0):
            zero.append(df.iloc[j][i])
        elif(j==1):
            sixty.append(df.iloc[j][i])
        elif(j==2):
            onetwenty.append(df.iloc[j][i])
        elif(j==3):
            oneeighty.append(df.iloc[j][i])
    
# Set position of bar on X axis
br1 = np.arange(len(zero))
br2 = [x + 1.3*barWidth for x in br1]
br3 = [x + 1.3*barWidth for x in br2]
br4 = [x + 1.3*barWidth for x in br3]

 
# Make the plot
plt.bar(br1, zero, color ='darkred', width = barWidth, edgecolor ='darkred', label ='0 mins')
plt.bar(br2, sixty, color ='orange', width = barWidth, edgecolor ='orange', label ='60 mins')
plt.bar(br3, onetwenty, color ='limegreen', width = barWidth, edgecolor ='limegreen', label ='120 mins')
plt.bar(br4, oneeighty, color ='red', width = barWidth, edgecolor ='red', label ='180 mins')


# Adding Xticks
plt.ylim(0, 0.9)
plt.xlabel('Cluster Size (n)', fontweight ='bold', fontsize = 15)
plt.ylabel('Number fraction', fontweight ='bold', fontsize = 15)
plt.xticks([r + 2*barWidth for r in range(len(zero))],['1', '2', '3', '4', '5', '>5'], fontsize=15)
plt.legend(fontsize = 15)
plt.yticks(fontsize=15)
plt.show()

#@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
```
